# Data Ingestion

Fetches sample CSV data from the internet and stores it in the RawData lakehouse using Polars.

```python
import polars as pl
from datetime import datetime
```

```python
# Fetch the penguins dataset from GitHub
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv"
df = pl.read_csv(url)
print(f"Loaded {len(df)} rows")
print(df.head())
```

```python
# Add metadata columns
df = df.with_columns([
    pl.lit(datetime.now()).alias("ingested_at"),
    pl.lit("seaborn-penguins").alias("source")
])
print("\nData with metadata:")
print(df.head())
```

```python
# Save to RawData lakehouse
onelake_path = "abfss://RawData@onelake.dfs.fabric.microsoft.com/Files/penguins.parquet"
df.write_parquet(onelake_path)
print(f"\nSaved {len(df)} rows to {onelake_path}")
```
